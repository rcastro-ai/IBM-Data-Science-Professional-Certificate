## Data Description

The dataset used for this visualization assignment contains *automobile_sales* data representing automobile sales and related variables during recession and non-recession period. 

The dataset includes the following variables:
<br>1. Date: The month end date of sales observation.
<br>2. Recession: A binary variable indicating recession perion; 1 means it was recession, 0 means it was normal.
<br>3. Automobile_Sales: The number of vehicles sold during the period.
<br>4. GDP: The per capita GDP value in USD.
<br>5. Unemployment_Rate: The monthly unemployment rate.
<br>6. Consumer_Confidence: A synthetic index representing consumer confidence, which can impact consumer spending and automobile purchases.
<br>7. Seasonality_Weight: The weight representing the seasonality effect on automobile sales during the period. This variable represents the seasonal effect on automobile sales for a given month. In the automobile industry, sales often fluctuate throughout the year due to seasonal patterns—for example, sales may increase during the festive season or year-end promotions, and decrease during off-peak months like post-holiday winter periods.
A value greater than 1 indicates higher-than-average sales expected for that month due to seasonal trends (e.g., holiday season or new model launches).
A value less than 1 suggests lower-than-average sales due to seasonal slowdowns.
A value around 1 means the season has neutral or average effect on sales.
<br>8. Price: The average vehicle price during the period.
<br>9. Advertising_Expenditure: The advertising expenditure of the company.
<br>10.Vehicle_Type: The type of vehicles sold; Supperminicar, Smallfamiliycar,                 Mediumfamilycar, Executivecar, Sports.
<br>11.Competition: The measure of competition in the market, such as the number of competitors or market share of major manufacturers.
<br>12.Month: Month of the observation extracted from Date..
<br>13.Year: Year of the observation extracted from Date.
<br>
By examining various factors mentioned above from the dataset, you aim to gain insights into how recessions impacted automobile sales for your company.

## Creating Visualizations for Data Analysis

### Showing a *Line Chart* using the functionality of pandas to show how Average automobile sales fluctuate from year to year

```python
# Creating a line chart showing how average automobile sales fluctuate from year to year

# Preparing the data
df_avg_sales = df.groupby('Year')['Automobile_Sales'].mean()

# Plotting the chart
plt.figure(figsize=(12, 8))

df_avg_sales.plot(x='Year', y='Automobile_Sales', kind='line')
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.title('Average Automobile Sales over Time')
plt.show()
```

### Now let's include the following on the plot

* Ticks on x- axis with all the years, to identify the years of recession 
* Aannotation for at least two years of recession
* Title as Automobile Sales during Recession


```python
# Creating a line chart showing how average automobile sales fluctuate from year to year showing ticks for all years and annotating the recession years

# Preparing the data
df_avg_sales = df.groupby('Year')['Automobile_Sales'].mean()

# Plotting the chart
plt.figure(figsize=(12, 8))

# Preparing ticks for x-axis
xticks = list(range(df['Year'].min(), df['Year'].max()+1))

ax = df_avg_sales.plot(x='Year', y='Automobile_Sales', kind='line', xticks=xticks, rot=90)
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.title('Average Automobile Sales over Time')

# Adding annotations
ax.annotate('Recession', 
            xy=(1991, 1400),       # The point to annotate (x, y)
            xytext=(1989, 5000),   # Where to place the text
            arrowprops=dict(facecolor='black', shrink=0.04)) # Adds an arrow
ax.annotate('Recession', 
            xy=(2001, 1400),       # The point to annotate (x, y)
            xytext=(1999, 5000),   # Where to place the text
            arrowprops=dict(facecolor='black', shrink=0.04)) # Adds an arrow

plt.show()

```

###  How do trends in advertising expenditure correlate with automobile sales during non-recession periods, and what insights can be derived from this relationship?

```python
# Create a regression plot to show how advertising expenditure and automobile sales correlate during onn-recession years

# Getting the data
df_adv_sales = df[df['Recession'] == 0]
df_adv_sales = df_adv_sales.groupby('Year')[['Advertising_Expenditure', 'Automobile_Sales']].mean().reset_index()

plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")

# Plotting two line plots
sns.lineplot(data=df_adv_sales, x='Year', y='Advertising_Expenditure', label='Average Advertising Expenditure', linestyle='--', color='blue', marker='s')
sns.lineplot(data=df_adv_sales, x='Year', y='Automobile_Sales', label='Average Automobile Sales', color='green', marker='o')

plt.title('Average Advertising Expenditure and Automobile Sales over Time')
plt.legend()   # Shows labels for both lines
plt.ylabel('') # Removing label next to y-axis
plt.show()
```

#### From the above plot, what insights do we have now?

**Analysis of Results**

From the chart we can see some kind of relationship between the average advertising expenditure and the average automobile sales during non-recession years though there are certain periods in which the behavior is quite different. This tends to indicate that advertising expenditure is a contributor to predicting automobile sales but there must be other features that can explain the variability.

### Using sub plotting let's compare the variations in GDP during recession and non-recession period by developing line plots for each period
Now, we want to find more insights from the data to understand the reason. 
Let's plot a two line charts using subplotting to answer:
#### How did the GDP vary over time during recession and non-recession periods? 

```python
# Using subplots to show GDP variations during recession and non-recession years

# Getting the data
df_non_rec = df[df['Recession'] == 0]
df_rec = df[df['Recession'] == 1]

# Create figure
fig = plt.figure(figsize=(16, 6)) 

ax0 = fig.add_subplot(1, 2, 1) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(1, 2, 2, sharey=ax0) # add subplot 2 (1 row, 2 columns, second plot)

# Subplot 1: Line plot for non-recession years
#df.plot(kind='line', figsize=(20, 6), ax=ax0) # add to subplot 1
sns.lineplot(x='Year', y='GDP', data=df_non_rec, label='Non Recession', ax=ax0)
ax0.set_title('GDP Variation During Non Recession Years')
ax0.set_xlabel('Year')
ax0.set_ylabel('GDP')

# Subplot 2: Line plot for recession years
#df_CI.plot(kind='line', figsize=(20, 6), ax=ax1) # add to subplot 2
sns.lineplot(x='Year', y='GDP', data=df_rec, label='Recession', ax=ax1)
ax1.set_title('GDP Variation During Recession Years')
ax1.set_xlabel('Year')
ax1.set_ylabel('GDP')

plt.tight_layout()
plt.show()
```

<details>
<summary>Inference</summary>
From this plot, it is evident that GDP tends to be lower and more volatile during recessions, with significant fluctuations and uncertainty. In contrast, non-recessionary periods show a relatively higher GDP level but still exhibit some degree of variability.<br><br>
</details>

### Creating a Bubble plot for displaying the impact of seasonality on Automobile Sales.
How has seasonality impacted the sales, in which months the sales were high or low? Let's focus on non-recession years to understand the trend.

#### Developing a Bubble plot for displaying Automobile Sales for every month and using Seasonality Weight for representing the size of each bubble

```python
# Plotting a bubble chart to show seasonality on automobile sales during non-recession years

# Getting the data
df_non_rec = df[df['Recession'] == 0]

weights = df_non_rec['Seasonality_Weight'] # For bubble size

# Plotting the chart
sns.scatterplot(data=df_non_rec, x='Month', y='Automobile_Sales', size=weights)

plt.xlabel('Month')
plt.ylabel('Automobile Sales')
plt.title('Seasonality Effect on Automobile Sales')

plt.show()
```

<details>
<summary>Inference</summary>
From this plot, it is evident that seasonality has not affected on the overall sales. However, there is a drastic raise in sales in the month of April and December<br><br></details>

 ### Creating a pie chart to display the portion of advertising expenditure of XYZAutomotives during recession and non-recession periods
 How did the advertising expenditure of XYZAutomotives change during recession and non-recession periods? 

```python
# Plotting a pie chart to show advertisment expenditure during recession and non-recession years

# Getting the data
df_rec = df[df['Recession'] == 1]
df_non_rec = df[df['Recession'] == 0]

# Getting totals
total_rec = df_rec['Advertising_Expenditure'].sum()
total_non_rec = df_non_rec['Advertising_Expenditure'].sum()

# Preparing data for pie chart
amounts = [total_rec, total_non_rec]
labels_list=['Recession', 'Non-Recession']
colors_list = ['lightgreen', 'tan']

# Plotting the chart
plt.figure(figsize=(8, 6))
plt.pie(amounts, labels=labels_list, colors=colors_list, autopct='%1.1f%%', startangle=90)

plt.title('Advertising Expenditure during Recession and Non-Recession Periods')
plt.show()
```

#### From the above plot, what insights do we find on the advertisement expenditure during recession and non recession periods?

**Analysis of Results**

The chart clearly shows that the advertising expenditure during non-recession years is four time larger than during recession years.

### Developing a pie chart to display the total Advertisement expenditure for each vehicle type during recession period

Can we observe the share of each vehicle type in total expenditure during recessions? 

```python
# Plotting a pie chart showing total advertising expenditure by vehicle type during recession periods

# Getting the data
df_rec = df[df['Recession'] == 1]

# Getting totals for advertising expenditure by vehicle type
total_rec_by_vtype = df_rec.groupby('Vehicle_Type')['Advertising_Expenditure'].sum()

# Plotting the chart
plt.figure(figsize=(8, 6))
plt.pie(total_rec_by_vtype.values, labels=total_rec_by_vtype.index, autopct='%1.1f%%', startangle=90)

plt.title('Share of Each Vehicle Type in Total Expenditure during Recessions')
plt.show()
```

### Creating a lineplot to analyse the effect of the unemployment rate on vehicle type and sales during the Recession Period
Let's analyze the effect of the unemployment rate on vehicle type and sales during the Recession Period.

```python
# Plotting a line chart showing the effect of the unemployment rate on automobile sales by vehicle type during recession periods

# Getting the data
df_rec = df[df['Recession'] == 1]

# Plotting the chart
plt.figure(figsize=(12, 6))

sns.lineplot(
    data=df_rec,
    x='unemployment_rate',               
    y='Automobile_Sales',
    hue='Vehicle_Type',
    marker='s'
)

# Adding title and labels
plt.title('Effect of Unemployment Rate on Sales by Vehicle Type')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Sales')
plt.legend(title='Vehicle Type')

# Show plot
plt.tight_layout()
plt.show()
```

Higher unemployment rates seem to be correlated to an important increase in the sales of super mini cars. Lower unemployment rates seem to suggest that favor the sales of medium family cars.

#### We found that the datset also contains the location/city for company offices. Now we want to show the recession impact on various offices/city sales by developing a choropleth.

```python
# Plotting a map showing the cities with the highest sales during recession periods

# Getting the data
df_rec = df[df['Recession'] == 1]

df_city_tot = df_rec.groupby('City')['Automobile_Sales'].sum().reset_index()

# Creating a map based on the U.S.
us_map = folium.Map(location=[37.0902, -95.7129], zoom_start=4)

# Create a choropleth layer using Folium
choropleth = folium.Choropleth(
             geo_data= 'us-states.json',  # GeoJSON file with state boundaries
             data=df_city_tot,
             columns=['City', 'Automobile_Sales'],
             key_on='feature.properties.name',
             fill_color='YlOrRd',
             fill_opacity=0.7,
             line_opacity=0.2,
             legend_name='Automobile Sales during Recession'
    ).add_to(us_map)

# Add tooltips to the choropleth layer
choropleth.geojson.add_child(
             folium.features.GeoJsonTooltip(['name'], labels=True)
)

# Display the map
us_map
```
