We could use `folium.Circle` to add a highlighted circle area with a text label on a specific coordinate. For example:

```python
# Creating a blue circle at NASA Johnson Space Center's coordinates with a popup label showing its name

circle = folium.Circle(nasa_coordinates, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('NASA Johnson Space Center'))

# Creating a blue circle at NASA Johnson Space Center's coordinates with an icon showing its name
marker = folium.map.Marker(
    nasa_coordinates,
    # Creating an icon as a text label
    icon = DivIcon(
        icon_size = (20,20),
        icon_anchor = (0,0),
        html = '<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'NASA JSC',
        )
    )
site_map.add_child(circle)
site_map.add_child(marker)
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get the coordinates for a mouse over a point on the map. As such, while we are exploring the map, we can easily find the coordinates of any points of interests (such as railways)

```python
# Adding Mouse Position to get the coordinates (Lat, Long) for a mouse over on the map

formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-right) in order to the distance to the launch site.

```python
# Creating a function to calculate the distance between two points on the map

from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
# Finding coordinates of the closest coastline

# Getting the coordinates for launch site: CCAFS SLC-40
launch_site_coordinates = spacex_df[spacex_df['Launch Site'] == 'CCAFS SLC-40'][['Lat', 'Long']].iloc[0].to_list()
launch_site_lat = launch_site_coordinates[0]
launch_site_lon = launch_site_coordinates[1]

# Latitude and longitude of coast point
coastline_lat = 28.56329
coastline_lon = -80.568

# Calculate distance in km
distance_coastline = round(calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon), 2)
```

```python
# Creating and adding a folium.Marker on the selected closest coastline point on the map
# Displaying the distance between coastline point and launch site using the icon property 

distance_marker = folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 14; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
        )
    )

site_map.add_child(distance_marker)
site_map
```

*TODO:* Draw a `PolyLine` between a launch site to the selected coastline point


```python
# Creating a `folium.PolyLine` object using the coastline coordinates and launch site coordinates

coordinates = [
    [launch_site_lat, launch_site_lon],      # Point A (Launch Site)
    [coastline_lat, coastline_lon]           # Point B (Coastline)
]
line = folium.PolyLine(locations=coordinates, weight=1)
site_map.add_child(line)
site_map
```

```python
# Creating a marker with distance to the closest city, railway, and highway. And drawing a line between the marker to the launch site

location_coordinates = [
    [28.08148, -80.61104],     # City
    [28.58212, -80.65434],     # Railway
    [28.56413, -80.57090]      # Highway
]

location_distances = []
for location in location_coordinates:
    # Calculating the distance between the location and the launch site
    location_lat = location[0]
    location_lon = location[1]
    location_distance = round(calculate_distance(launch_site_lat, launch_site_lon, location_lat, location_lon), 2)
    location_distances.append(location_distance)

    # Creating the marker for the location
    location_marker = folium.Marker(
        [location_lat, location_lon],
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 14; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(location_distance),
            )
        )

    # Adding marker to map
    site_map.add_child(location_marker)

    # Adding line to map
    line = folium.PolyLine(locations=[[launch_site_lat, launch_site_lon], [location_lat, location_lon]], weight=1)
    site_map.add_child(line)

site_map
```
